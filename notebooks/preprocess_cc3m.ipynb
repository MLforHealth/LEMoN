```python
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm

img_dir = Path('/data/healthy-ml/gobi1/data/cc3m-wds/cc3m-wds/train')
```

```python
raw = []
for i in tqdm(img_dir.glob('**/*.jpg')):
    fname = i.with_suffix('').name
    label_name = i.parent/(fname+'.txt')
    if not label_name.is_file():
        continue
    label = label_name.open('r').read()
    raw.append({
        'path': str(i),
        'label': label
    })
```

```python
df = pd.DataFrame(raw)
df['split'] = 'train'
```

```python
df = df.rename(columns = {
    'label': 'sentence'
})
```

```python
df['gold_sentence'] = ''
```

```python
df.to_pickle('/data/healthy-ml/gobi1/data/cc3m-wds/cc3m-wds/multimodal_mislabel_split.pkl')
```
